```python
import pandas as pd
import geopandas as gpd
from shapely.wkt import loads
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/london-2021-lsoa-gen20.csv')

lsoa_gdf = gpd.GeoDataFrame(df)
lsoa_gdf['geometry'] = lsoa_gdf['geometry'].apply(lambda x: loads(x))
lsoa_gdf = lsoa_gdf.set_geometry('geometry')

print(lsoa_gdf.shape)
print(lsoa_gdf.columns.tolist())
```

```python
pm25_df = pd.read_csv('data/lsoa21-2019-PM25.csv')

print(pm25_df.shape)
print(pm25_df.columns.tolist())
print(pm25_df.head())
```

```python
gdf = lsoa_gdf.merge(pm25_df, left_on='LSOA21CD', right_on='LSOA21CD', how='left')

print(gdf.shape)
print(gdf.head())
```

```python
print(gdf['PM25-2019-mean'].isna().sum())
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 10))

gdf.plot(column='PM25-2019-mean',
         cmap='YlOrRd',
         scheme='quantiles',
         k=5,
         legend=True,
         legend_kwds={
             'title': 'PM2.5 (µg/m³)',
             'loc': 'lower left'
         },
         ax=ax)
legend = ax.get_legend()
for text in legend.get_texts():
    text.set_text(text.get_text().replace(',', ' -'))
    
ax.set_title('Mean annual PM2.5 concentration across London LSOAs (2019)',
             fontsize=16, pad=15, fontweight='bold')
ax.set_axis_off()

plt.tight_layout()
plt.savefig('fig1_pm25_choropleth.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from libpysal import weights
from pysal.explore import esda
from splot import esda as esdaplot
```

```python
w = weights.Queen.from_dataframe(gdf)
w.transform = 'R'  

print(w.n)          
print(w.pct_nonzero)  
```

```python
moran = esda.Moran(gdf['PM25-2019-mean'], w)

print(f"Moran's I: {moran.I:.4f}")
print(f"Expected I: {moran.EI:.4f}")
print(f"p-value: {moran.p_sim:.4f}")
print(f"z-score: {moran.z_norm:.4f}")
```

```python
from esda.moran import Moran
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(8, 8))

lag = weights.spatial_lag.lag_spatial(w, gdf['PM25-2019-mean'])

pm25_std = (gdf['PM25-2019-mean'] - gdf['PM25-2019-mean'].mean()) / gdf['PM25-2019-mean'].std()
lag_std = (lag - lag.mean()) / lag.std()

sc = ax.scatter(pm25_std, lag_std,
                c=gdf['PM25-2019-mean'],
                cmap='YlOrRd',
                edgecolor='none',
                alpha=0.5,
                s=10)

m, b = np.polyfit(pm25_std, lag_std, 1)
ax.plot(pm25_std.sort_values(),
        m * pm25_std.sort_values() + b,
        color='black', linewidth=1.5)

ax.axhline(0, color='grey', linewidth=0.8, linestyle='--')
ax.axvline(0, color='grey', linewidth=0.8, linestyle='--')

plt.colorbar(sc, ax=ax, label='PM2.5 (µg/m³)', shrink=0.6)

ax.set_title("Moran's I scatterplot — PM2.5 across London LSOAs (2019)",
             fontsize=13, fontweight='bold', pad=15)
ax.set_xlabel('PM2.5 (standardised)', fontsize=11)
ax.set_ylabel('Spatial lag of PM2.5 (standardised)', fontsize=11)

ax.annotate(f"Moran's I = {moran.I:.4f}\np = {moran.p_sim:.4f}",
            xy=(0.05, 0.95), xycoords='axes fraction',
            fontsize=11, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.savefig('fig2_morans_i_scatterplot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
lisa = esda.Moran_Local(gdf['PM25-2019-mean'], w)

print(f"Number of significant LSOAs (p < 0.05): {(lisa.p_sim < 0.05).sum()}")
```

```python
gdf['lisa_q'] = lisa.q          
gdf['lisa_p'] = lisa.p_sim    
gdf['lisa_sig'] = lisa.p_sim < 0.05   
```

```python
conditions = [
    (gdf['lisa_sig'] == True) & (gdf['lisa_q'] == 1),
    (gdf['lisa_sig'] == True) & (gdf['lisa_q'] == 2),
    (gdf['lisa_sig'] == True) & (gdf['lisa_q'] == 3),
    (gdf['lisa_sig'] == True) & (gdf['lisa_q'] == 4),
]
choices = ['High-High', 'Low-High', 'Low-Low', 'High-Low']

gdf['lisa_label'] = np.select(conditions, choices, default='Not significant')

print(gdf['lisa_label'].value_counts())
```

```python
import numpy as np
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch

colour_map = {
    'High-High': '#e2191c',      
    'Low-Low': '#ffffcc',        
    'High-Low': '#fd8c3c',       
    'Low-High': '#800026',       
    'Not significant': '#d9d9d9' 
}

gdf['colour'] = gdf['lisa_label'].map(colour_map)

fig, ax = plt.subplots(1, 1, figsize=(12, 10))

gdf.plot(color=gdf['colour'],
         linewidth=0.1,
         edgecolor='white',
         ax=ax)

legend_elements = [
    Patch(facecolor='#ffffcc', label='Low-Low (coldspot)'),
    Patch(facecolor='#fd8c3c', label='High-Low (outlier)'),
    Patch(facecolor='#e2191c', label='High-High (hotspot)'),
    Patch(facecolor='#800026', label='Low-High (outlier)'),
    Patch(facecolor='#d9d9d9', label='Not significant')
]

ax.legend(handles=legend_elements,
          loc='lower left',
          title='LISA cluster type',
          title_fontsize=10,
          fontsize=9,
          framealpha=0.9)

ax.set_title('LISA cluster map — PM2.5 across London LSOAs (2019)',
             fontsize=16, fontweight='bold', pad=15)
ax.set_axis_off()

plt.tight_layout()
plt.savefig('fig3_lisa_cluster_map.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
imd_df = pd.read_csv('data/IMD 2019-Table 1.csv')

imd_df = imd_df[['LSOA code (2011)', 'Index of Multiple Deprivation (IMD) Score']]
imd_df.columns = ['LSOA21CD', 'IMD_score']

gdf = gdf.merge(imd_df, on='LSOA21CD', how='left')

print(f"Total LSOAs: {len(gdf)}")
print(f"LSOAs with IMD data: {gdf['IMD_score'].notna().sum()}")
print(f"LSOAs missing IMD data: {gdf['IMD_score'].isna().sum()}")
```

```python

gdf_biv = gdf.dropna(subset=['IMD_score']).copy()

print(f"LSOAs for bivariate analysis: {len(gdf_biv)}")
```

```python
w_biv = weights.Queen.from_dataframe(gdf_biv)
w_biv.transform = 'R'

print(f"Weights matrix built for {w_biv.n} LSOAs")
```

```python
from esda.moran import Moran_BV

moran_bv = Moran_BV(gdf_biv['PM25-2019-mean'], 
                     gdf_biv['IMD_score'], 
                     w_biv)

print(f"Bivariate Moran's I: {moran_bv.I:.4f}")
print(f"p-value: {moran_bv.p_sim:.4f}")
print(f"z-score: {moran_bv.z_sim:.4f}")
```

```python
lisa_bv = esda.Moran_Local_BV(gdf_biv['PM25-2019-mean'],
                                gdf_biv['IMD_score'],
                                w_biv)

gdf_biv['bv_q'] = lisa_bv.q
gdf_biv['bv_p'] = lisa_bv.p_sim
gdf_biv['bv_sig'] = lisa_bv.p_sim < 0.05

conditions = [
    (gdf_biv['bv_sig'] == True) & (gdf_biv['bv_q'] == 1),
    (gdf_biv['bv_sig'] == True) & (gdf_biv['bv_q'] == 2),
    (gdf_biv['bv_sig'] == True) & (gdf_biv['bv_q'] == 3),
    (gdf_biv['bv_sig'] == True) & (gdf_biv['bv_q'] == 4),
]
choices = ['High PM2.5 - High IMD', 'Low PM2.5 - High IMD',
           'Low PM2.5 - Low IMD', 'High PM2.5 - Low IMD']

gdf_biv['bv_label'] = np.select(conditions, choices, 
                                  default='Not significant')

print(gdf_biv['bv_label'].value_counts())
```

```python
from matplotlib.patches import Patch

colour_map_bv = {    
    'Low PM2.5 - Low IMD':   '#ffffcc',    
    'Low PM2.5 - High IMD':  '#fdae44',    
    'High PM2.5 - Low IMD':  '#e2191c',  
    'High PM2.5 - High IMD': '#800026',
    'Not significant':       '#d9d9d9'    
}

gdf_biv['colour'] = gdf_biv['bv_label'].map(colour_map_bv)

fig, ax = plt.subplots(1, 1, figsize=(12, 10))

gdf_biv.plot(color=gdf_biv['colour'],
             linewidth=0.1,
             edgecolor='white',
             ax=ax)

legend_elements = [
    Patch(facecolor='#ffffcc', label='Low PM2.5 - Low IMD'),
    Patch(facecolor='#fdae44', label='Low PM2.5 - High IMD'),
    Patch(facecolor='#e2191c', label='High PM2.5 - Low IMD'),
    Patch(facecolor='#800026', label='High PM2.5 - High IMD'),
    Patch(facecolor='#d9d9d9', label='Not significant')
    
]

ax.legend(handles=legend_elements,
          loc='lower left',
          title='Bivariate LISA cluster type',
          title_fontsize=10,
          fontsize=9,
          framealpha=0.9)

ax.set_title('Bivariate LISA — PM2.5 and deprivation across London LSOAs (2019)',
             fontsize=16, fontweight='bold', pad=15)
ax.set_axis_off()

plt.tight_layout()
plt.savefig('fig4_bivariate_lisa.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from scipy import stats

fig, ax = plt.subplots(1, 1, figsize=(8, 6))

ax.scatter(gdf_biv['IMD_score'], 
           gdf_biv['PM25-2019-mean'],
           alpha=0.3, 
           color='#d7301f',
           edgecolor='none',
           s=8)

slope, intercept, r, p, se = stats.linregress(gdf_biv['IMD_score'], 
                                               gdf_biv['PM25-2019-mean'])
x_line = np.linspace(gdf_biv['IMD_score'].min(), 
                     gdf_biv['IMD_score'].max(), 100)
ax.plot(x_line, slope * x_line + intercept, 
        color='black', linewidth=1.5)

ax.annotate(f'r = {r:.3f}, p < 0.001',
            xy=(0.05, 0.95), xycoords='axes fraction',
            fontsize=11, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

ax.set_xlabel('IMD Score (higher = more deprived)', fontsize=11)
ax.set_ylabel('Mean PM2.5 (µg/m³)', fontsize=11)
ax.set_title('Relationship between deprivation and PM2.5 across London LSOAs (2019)',
             fontsize=12, fontweight='bold', pad=15)

plt.tight_layout()
plt.savefig('fig4_scatter_imd_pm25.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"r = {r:.4f}")
print(f"r² = {r**2:.4f}")
print(f"p = {p:.4f}")
```
